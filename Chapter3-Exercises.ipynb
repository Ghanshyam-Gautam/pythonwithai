```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load the dataset
df = pd.read_csv("weight-height.csv")

# Step 2: Extract height and weight
height = df["Height"].values.reshape(-1, 1)
weight = df["Weight"].values

# Step 3: Scatter plot
plt.scatter(height, weight, alpha=0.5)
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.title("Height vs Weight")
plt.show()

# Step 4: Fit linear regression
model = LinearRegression()
model.fit(height, weight)
predicted_weight = model.predict(height)

# Step 5: Plot regression line
plt.scatter(height, weight, alpha=0.5)
plt.plot(height, predicted_weight, color='red')
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.title("Linear Regression: Height vs Weight")
plt.show()

# Step 6: Evaluate model
import numpy as np
rmse = np.sqrt(mean_squared_error(weight, predicted_weight))
r2 = r2_score(weight, predicted_weight)

print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
Visual Assessment:
The scatter plot shows a clear upward trend, and the red regression line fits the data well. Most data points are close to the line, indicating a good match between predicted and actual values.
Numerical Assessment: 
R² Score = 0.8552 → This means that 85.52% of the variation in weight is explained by height. It shows a strong linear relationship.
RMSE = 12.22 → This is the average error in predicted weight, which is relatively low considering the scale of the data.
```
