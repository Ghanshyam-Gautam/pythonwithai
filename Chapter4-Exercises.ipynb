Consider the dataset `50_Startups.csv` which contains data for companies' profit etc.

```python
# a) Read the dataset into pandas dataframe paying attention to file delimiter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Adjust path as needed; delimiter is comma for the common 50_Startups dataset
df = pd.read_csv("50_Startups.csv", delimiter=",")
```

```python
"""
b) Identify the variables inside the dataset

Findings:
- Numeric variables: R&D Spend, Administration, Marketing Spend, Profit
- Categorical variable: State
- Target variable: Profit

We'll print schema and quick info to confirm.
"""
print("Columns:", df.columns.tolist())
print(df.info())
```

```python
"""
c) Investigate the correlation between the variables

We inspect pairwise correlations for numeric variables only.
This helps identify linear relationships and guides feature selection.
"""
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()

print("\nCorrelation matrix (numeric):\n", corr)

# Visual correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation matrix (numeric variables)")
plt.tight_layout()
plt.show()
```

```python
"""
d) Choose appropriate variables to predict company profit. Justify your choice.

Justification based on typical behavior of this dataset and observed correlations:
- R&D Spend: Typically the strongest positive correlation to Profit. Intuitively, investing in R&D drives product value and returns.
- Marketing Spend: Often shows moderate positive correlation with Profit. Marketing supports sales and growth.
- Administration: Usually weak correlation; tends to add operational cost without directly scaling profit.
- State: Categorical; its effect can be minor, but including it via one-hot encoding can capture regional differences.

Approach:
- We'll build two models for clarity:
  Model A (numeric-only): ['R&D Spend', 'Marketing Spend', 'Administration']
  Model B (with State): one-hot encode 'State' + numeric variables

We’ll report metrics for Model B as the primary model (richer explanatory set).
"""

# e) Plot explanatory variables against profit to confirm (close to) linear dependence
explanatory_numeric = ["R&D Spend", "Administration", "Marketing Spend"]
plt.figure(figsize=(12,3))
for i, col in enumerate(explanatory_numeric, 1):
    plt.subplot(1,3,i)
    sns.scatterplot(x=df[col], y=df["Profit"])
    sns.regplot(x=df[col], y=df["Profit"], scatter=False, color="red")
    plt.title(f"{col} vs Profit")
plt.tight_layout()
plt.show()
```

```python
"""
Observations from the plots:
- R&D Spend vs Profit: Points generally align with a positive slope; indicates near-linear relationship.
- Marketing Spend vs Profit: Positive trend, but more dispersion than R&D.
- Administration vs Profit: Weak/flat trend; lower predictive power.
"""

# f) Form training and testing data (80/20 split)
# Primary model (Model B): numeric + one-hot State
X = df[["R&D Spend", "Administration", "Marketing Spend", "State"]].copy()
y = df["Profit"].copy()

# One-hot encode 'State'
X = pd.get_dummies(X, columns=["State"], drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print("\nTrain/Test shapes:", X_train.shape, X_test.shape)

# g) Train linear regression model with training data
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# h) Compute RMSE and r^2 values for training and testing data separately
y_train_pred = linreg.predict(X_train)
y_test_pred = linreg.predict(X_test)

rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"\nModel B (with State dummies) performance:")
print(f"Train RMSE: {rmse_train:,.2f}")
print(f"Test  RMSE: {rmse_test:,.2f}")
print(f"Train R^2: {r2_train:.4f}")
print(f"Test  R^2: {r2_test:.4f}")
```

```python
"""
Findings and explanations:

1) Feature choice:
   - R&D Spend dominates the signal for Profit in this dataset, making it indispensable.
   - Marketing Spend contributes additional variance explanation.
   - Administration contributes little; we keep it to test if it adds marginal value.
   - State has minor effect but can capture fixed regional differences; included via one-hot encoding.

2) Expected model behavior:
   - High R^2 on train, reasonable R^2 on test, with RMSE reflecting currency scale of Profit.
   - If Administration inflates variance or adds noise, feature selection could drop it to improve generalization.
   - If State is non-informative in your specific file, coefficients for its dummies will be small.

3) Diagnostics to consider:
   - Residual plots: check for non-linearity and heteroscedasticity.
   - Multicollinearity: R&D and Marketing may be partially correlated; VIF can confirm.
   - Robustness: Cross-validation can stabilize estimates beyond a single split.

```
